```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('employer.csv')
df.head()
```

```python
df.head()
```

```python
#Removing Unwanted colemns
data1 = df.drop(columns=['parameters','Employeer_employee','IMP_parameters','cds','motivationg','waysmotivation','conflicts','wcs','s'])
```

```python
#One Hot Encoding Part
data1 = pd.get_dummies(data1, columns = ['weorke','communication','conflict'])
```

```python
data1.head()
```

```python
data2 = data1.iloc[:,0:8]
```

# Finding The Correlation between the variables

```python
import seaborn as sns
import numpy as np
sns.set(style="white")
#Plotting the correlation between every variables
corr = data2.corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(270, 9, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

# Plotting the Variables
The pairs plot builds on two basic figures, the histogram and the scatter plot. The histogram on the diagonal allows us to see the distribution of a single variable while the scatter plots on the upper and lower triangles show the relationship (or lack thereof) between two variables. For example, the left-most plot in the second row shows the scatter plot of life_exp versus year.

# Histogram Distribution of Variables


```python
from collections import Counter
df1 = pd.read_csv('employer.csv')
list1 = ['parameters', 'communication', 'weorke', 'conflict']
for i in range(4):
    letter_counts = Counter(df1[list1[i]])
    df = pd.DataFrame.from_dict(letter_counts, orient='index')
    ax = df.plot(kind='bar', title = ('Frequency plot of '+list1[i]))
    ax.set_xlabel(list1[i])
    ax.set_ylabel("Frequency")
```
